```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# Curve Plotting

```python
dset.plot(kind='scatter', x = 'sqft_living', y = 'price')
dset.plot(kind='scatter', x = 'bedrooms', y = 'price')
dset.plot(kind='scatter', x = 'bathrooms', y = 'price')
dset.plot(kind='scatter', x = 'floors', y = 'price')
plt.show()
```

```python
dset.corr()
```

```python
price = pd.DataFrame(dset['price'])
sqft_living = pd.DataFrame(dset['sqft_living'])
bedrooms = pd.DataFrame(dset['bedrooms'])
bathrooms = pd.DataFrame(dset['bathrooms'])
floors = pd.DataFrame(dset['floors'])
```

# Simple Linear_Regression_Model_Creation

```python
lm = linear_model.LinearRegression()
model1 = lm.fit(price,sqft_living)
```

```python
model1.intercept_
```

```python
model1.score(price,sqft_living)
```

```python
model2 = lm.fit(price,bedrooms)
model2.intercept_
```

```python
model2.score(price,bedrooms)
```

```python
model3 = lm.fit(price,bathrooms)
model3.intercept_
```

```python
model3.score(price,bathrooms)
```

```python
model4 = lm.fit(price,floors)
model4.intercept_
```

```python
model4.score(price,floors)
```

# Multi_Linear_Regression_Model_Creation

```python
y = pd.DataFrame(dset['price'])
X = pd.DataFrame(dset[['sqft_living','bedrooms','bathrooms','floors']])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
multi_reg= lm.fit(X_train,y_train)
```

```python
y_pred = multi_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
multi_reg.coef_
```

```python
multi_reg.intercept_
```

1. Multi_Regression Model has been created with 50.63% accuracy.
2. To Increase accuracy Pre-processing techniques like, Removal of Outliers can be adopted on the Data.
3.Bedrooms & floor have least score in Price value while doing Simple Regression.
