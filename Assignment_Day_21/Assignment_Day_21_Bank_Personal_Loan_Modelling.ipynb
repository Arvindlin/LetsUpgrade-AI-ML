```python
import pandas as pd
import statsmodels.api as sm
```

```python
d_set1= pd.read_excel("Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
```

```python
d_set1.head()
```

```python
d_set1=d_set1.drop(columns=['ID', 'ZIP Code'])
```

```python
d_set1.head(5)
```

# Creating Dependent Variable and Independent Variable.

```python
Y = d_set1[['Personal Loan']]
X = d_set1[['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','Online','CreditCard']]
```

# Defining the Constant.

```python
X1 = sm.add_constant(X)
```

# Creating Logistic Model

```python
Logi = sm.Logit(Y,X1)
result = Logi.fit()
```

# Creating Summary of Model

```python
result.summary()
```

# Conclusion from above Summary:-
1. Mortgage has very low significanc, on weather the Customer will take the Personal Loan or not from the Bank as p-value is greater than 0.05.
2. All other Independent Varable except "Mortgage" have high significant, on weather the Customer will take the Personal Loan or not from the Bank.
