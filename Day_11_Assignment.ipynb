```python
import pandas as pd
from scipy.stats import pearsonr as ps
dataset = pd.read_csv("general_data.csv")
dataset.head()
dataset.dropna()     # To drop Null values
dataset.drop_duplicates()     # To drop duplicate values
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
dataset["Attrition"] = lb.fit_transform(dataset["Attrition"])
dataset["BusinessTravel"] = lb.fit_transform(dataset["BusinessTravel"])
dataset["Department"] = lb.fit_transform(dataset["Department"])
dataset["EducationField"] = lb.fit_transform(dataset["EducationField"])
dataset["Gender"] = lb.fit_transform(dataset["Gender"])
dataset["Over18"] = lb.fit_transform(dataset["Over18"])
dataset.head(5)
```

```python
stats,p= ps(dataset.Attrition,dataset.Age)
print("Correlation of Attrition with Age is =",stats)

stats,p= ps(dataset.Attrition,dataset.BusinessTravel)
print("Correlation of Attrition with BusinessTravel is =",stats)

stats,p= ps(dataset.Attrition,dataset.Department)
print("Correlation of Attrition with Department is =",stats)

stats,p= ps(dataset.Attrition,dataset.DistanceFromHome)
print("Correlation of Attrition with DistanceFromHome is =",stats)

stats,p= ps(dataset.Attrition,dataset.Education)
print("Correlation of Attrition with Education is =",stats)

stats,p= ps(dataset.Attrition,dataset.EducationField)
print("Correlation of Attrition with EducationField is =",stats)

stats,p= ps(dataset.Attrition,dataset.Gender)
print("Correlation of Attrition with Gender is =",stats)


stats,p= ps(dataset.Attrition,dataset.PercentSalaryHike)
print("Correlation of Attrition with PercentSalaryHike is =",stats)

stats,p= ps(dataset.Attrition,dataset.StockOptionLevel)
print("Correlation of Attrition with StockOptionLevel is =",stats)


stats,p= ps(dataset.Attrition,dataset.TrainingTimesLastYear)
print("Correlation of Attrition with TrainingTimesLastYear is =",stats)

stats,p= ps(dataset.Attrition,dataset.YearsAtCompany)
print("Correlation of Attrition with YearsAtCompany is =",stats)

stats,p= ps(dataset.Attrition,dataset.YearsSinceLastPromotion)
print("Correlation of Attrition with YearsSinceLastPromotion is =",stats)

stats,p= ps(dataset.Attrition,dataset.YearsWithCurrManager)
print("Correlation of Attrition with YearsWithCurrManager is =",stats)
```
